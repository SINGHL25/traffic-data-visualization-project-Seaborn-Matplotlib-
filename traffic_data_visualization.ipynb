```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("traffic_data.csv")  # Ensure CSV is in same folder or adjust path

# Set seaborn style
sns.set(style="whitegrid")

# 1. Barplot: Vehicle Class Count
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='VehicleClass')
plt.title('Vehicle Class Distribution')
plt.show()

# 2. Histogram: Speed
plt.figure(figsize=(8, 5))
sns.histplot(df['Speed'], bins=10, kde=True)
plt.title('Speed Distribution')
plt.show()

# 3. Pie Chart: ANPR Status
df['ANPR_Status'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('ANPR Status Distribution')
plt.ylabel('')
plt.show()

# 4. Box Plot: Speed by Vehicle Class
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='VehicleClass', y='Speed')
plt.title('Box Plot of Speed by Vehicle Class')
plt.show()

# 5. Swarm Plot: Speed by ANPR Status
plt.figure(figsize=(8, 6))
sns.swarmplot(data=df, x='ANPR_Status', y='Speed')
plt.title('Swarm Plot of Speed by ANPR Status')
plt.show()

# 6. Strip Plot: Speed by TollPoint
plt.figure(figsize=(8, 6))
sns.stripplot(data=df, x='TollPoint', y='Speed', jitter=True)
plt.title('Strip Plot of Speed by TollPoint')
plt.show()

# 7. Pair Plot: Speed, PassageCount, TransactionCount
sns.pairplot(df[['Speed', 'PassageCount', 'TransactionCount']])
plt.suptitle('Pair Plot of Traffic Metrics', y=1.02)
plt.show()

# 8. Heatmap: Correlation
plt.figure(figsize=(8, 6))
corr = df[['Speed', 'PassageCount', 'TransactionCount']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# 9. Joint Plot: Speed vs TransactionCount
sns.jointplot(data=df, x='Speed', y='TransactionCount', kind='reg')
plt.suptitle('Joint Plot with Regression', y=1.02)
plt.show()

# 10. Facet Grid: Speed Distribution by TollPoint
g = sns.FacetGrid(df, col='TollPoint')
g.map(sns.histplot, 'Speed', bins=10)
g.fig.suptitle('Speed Distribution by TollPoint', y=1.02)
plt.show()

# 11. Scatter Plot: PassageCount vs TransactionCount
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='PassageCount', y='TransactionCount', hue='VehicleClass', style='ANPR_Status')
plt.title('Scatter Plot of Passage vs Transaction Count')
plt.show()
```
